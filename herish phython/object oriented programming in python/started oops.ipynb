```python
#how we create own data type
class Fraction:
    def __init__(self,num,den):#this is an constructor
        self.num=num
        self.den=den
    def __str__(self):#for print any string this is an magic method in python 
        return "{}/{}".format(self.num,self.den)
    def __add__(self,other):#it is use to show add functionality and help to print
        temp_num=self.num*other.den+other.num*self.den
        temp_den=self.den*other.den
        return "{}/{}".format(temp_num,temp_den)
    def __sub__(self,other):#it is use to show subtract functionality and help to print
        temp_num=self.num*other.den-other.num*self.den
        temp_den=self.den*other.den
        return "{}/{}".format(temp_num,temp_den)
    def __truediv__(self,other):#it is use to show div functionality and help to print
        temp_num=self.num*other.num
        temp_den=self.den*other.den
        return "{}/{}".format(temp_num,temp_den)
    def __mul__(self,other):#it is use to show multiplication functionality and help to print
        temp_num=self.num*other.num
        temp_den=self.den*other.den
        return "{}/{}".format(temp_num,temp_den)
x=Fraction(10,5)       
y=Fraction(9,5)
print(y)
print(x)
print(x+y)
print(x-y)
print(x/y)
print(x*y)
        
```
